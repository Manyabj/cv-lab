```python
from PIL import Image  # Importing the Image module from the PIL library
import numpy as np    # Importing numpy for array manipulation
import matplotlib.pyplot as plt  # Importing matplotlib for plotting

img = Image.open('C:\\Users\\sinch\\Downloads\\image1.webp')  # Open the image file
img = np.asarray(img)  # Convert the image to a numpy array
plt.imshow(img)  # Display the image
plt.title('Original Image', c='r')  # Set the title of the plot
plt.show()  # Show the plot
```

```python
# Create a horizontally mirrored version of the image by reversing the order of columns
hori = img[:, ::-1, :]

# Display the horizontally mirrored image using matplotlib
plt.imshow(hori)

# Set the title of the plot with green color
plt.title('Horizontally Mirrored Image', c='g')

# Show the plot
plt.show()
```

```python
# Flip the image both vertically and horizontally
flipped = img[::-1, ::-1, :]

# Display the flipped image using matplotlib
plt.imshow(flipped)

# Set the title of the plot with green color
plt.title('Double Flipped Image', c='g')

# Show the plot
plt.show()
```
